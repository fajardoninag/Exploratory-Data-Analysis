```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", hue="sex", data=titanic)

plt.title("Average Fare by Passenger Class and Gender", fontsize=14, fontweight="bold")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.scatterplot(x="age", y="fare", hue="survived", data=titanic)
plt.title("Age vs. Fare with Survival Status")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

correlation_matrix = titanic.select_dtypes(include=["number"]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))


bar_custom = sns.barplot( x="class", y="fare", hue="sex", data=titanic,
    palette={"male": "#ADD8E6", "female": "#FFB6C1"},  edgecolor="black",errorbar=None )


for p in bar_custom.patches:
    height = p.get_height()
    if not height:
        continue
    bar_custom.annotate(
        f'{height:.2f}',
        (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', fontsize=10, color='black')


plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Average Fare", fontsize=12)
plt.title("Average Fare by Passenger Class and Gender", fontsize=14, fontweight="bold")

plt.legend(title="Gender", loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
custom_palette = {0: "#D3D3D3", 1: "#32CD32"}

plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
scatter = sns.scatterplot(x="age", y="fare", hue="survived", data=titanic,
    palette=custom_palette, alpha=0.7)


plt.xlabel("Age", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.title("Age vs. Fare with Survival Status", fontsize=14, fontweight="bold")

handles, labels = scatter.get_legend_handles_labels()

legend_labels= ["0", "1"]
plt.legend(handles=handles, labels=legend_labels, title="Survival Status", loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")
correlation_matrix = titanic.select_dtypes(include=["number"]).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,cmap="Blues",linewidths=0.5, linecolor="white", vmin=-1, vmax=1)

plt.title(" Correlation Matrix of Titanic", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

plt.show()
```
