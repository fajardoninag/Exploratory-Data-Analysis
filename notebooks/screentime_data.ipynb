```python
from IPython import get_ipython
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np




data = pd.read_csv("screentime_analysis.csv")



times_opened = data['Times Opened']
usage_time = data['Usage (minutes)']

norm = plt.Normalize(usage_time.min(), usage_time.max()) #the normalization function is used to map a given number to a given range (0,1 in this case)

cmap = cm.get_cmap('Greens') #select the colormap


colors = cmap(norm(usage_time)) #get the colors by using the colormap function with the normalized data.

   # Create the scatter plot with colors
plt.scatter(times_opened, usage_time, color=colors, edgecolors="black", linewidths=1, alpha=0.75) #the c is used to assign the colors
cbar= plt.colorbar()


# Create a scatter plot

plt.xlabel('Times Opened')
plt.ylabel('Usage Time in minutes')
plt.title('Relationship between Scatter Plot of Times Opened and Usage Time')

plt.tight_layout

# Show plot
plt.show()


```

```python
application = data['App']
usage_time = data['Usage (minutes)']


plt.plot( usage_time,application,linestyle='--',color= 'pink')
plt.xlabel('Notifications')
plt.ylabel('Application')
plt.title('Highest Notified App')
plt.xticks(rotation=30)

plt.show()
```

```python

application = data['App']
notification = data['Notifications']

plt.bar(application, notification, width=0.8, color= 'pink', alpha=1)
plt.xlabel('Application')
plt.ylabel('Notification')
plt.title('Highest Notified App')
plt.xticks(rotation=30)

plt.show()
```
