```python
# Subplots
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# Percentage
survival_gender_class = pd.crosstab([df['sex'], df['pclass']], df['survived'], normalize='index') * 100

# First plot
survival_gender_class.plot(kind='bar', stacked=True, ax=ax[0], color=['lightcoral', 'lightgreen'])
ax[0].set_title('Survival Rate by Gender and Passenger Class')
ax[0].set_xlabel('Gender and Passenger Class')
ax[0].set_ylabel('Survival Rate')
ax[0].tick_params(axis='x', rotation=45)
ax[0].legend(labels=['Did Not Survive', 'Survived'], loc='upper right', bbox_to_anchor=(1, 1))

# Percentage
age_group_survival = pd.crosstab(df['age_group'], df['survived'], normalize='index') * 100

# Second plot
age_group_survival.plot(kind='bar', stacked=True, ax=ax[1], color=['lightcoral', 'lightgreen'])
ax[1].set_title('Survival Rate by Age ')
ax[1].set_xlabel('Age Group')
ax[1].set_ylabel('Survival Rate')
ax[1].tick_params(axis='x', rotation=45)
ax[1].legend(labels=['Did Not Survive', 'Survived'], loc='upper right', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

Do age, class, and gender affect the survival rate?

•	Survival Rate by Gender and Class

-	Categorical vs numerical analysis
•	Survival Rate by Age: stacked bar plot
categorical - gender and class, numerical - age


```python
# Create a new feature family size
df['family_size'] = df['sibsp'] + df['parch'] + 1


# Create a cross-tab
family_survival = pd.crosstab(df['family_size'], df['survived'])

# Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(family_survival, annot=True, fmt='d', cmap='coolwarm')
plt.title('Survival Count by Family Size')
plt.xlabel('Survival')
plt.ylabel('Family Size')
plt.show()



```

Does the family size influence the survival rate of the passenger?
