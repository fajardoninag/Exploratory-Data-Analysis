```python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
file_path = "cleaned_data.csv"  # Update this path if needed
df = pd.read_csv(file_path)

# Define strands in the requested order
strands_dict = {
    "acad_-_humss": "HUMSS",
    "acad_-_abm": "ABM",
    "acad_stem": "STEM",
    "acad_gas": "GAS",
    "tvl": "TVL",
    "sports": "SPORTS",
    "arts": "ARTS",
    "acad_pbm": "PBM"
}

# Define nodes
labels = ["Grade 11", "Grade 12", "Male", "Female"] + list(strands_dict.values())

# Assign numeric indices
grade_11_index = labels.index("Grade 11")
grade_12_index = labels.index("Grade 12")
male_index = labels.index("Male")
female_index = labels.index("Female")

# Initialize link data
source = []
target = []
values = []

# Process data for Sankey
for strand in strands_dict:
    strand_index = labels.index(strands_dict[strand])

    # Add connections for Grade 11
    g11_male = df[f"g11_{strand}_male"].sum()
    g11_female = df[f"g11_{strand}_female"].sum()
    source.extend([grade_11_index, grade_11_index])
    target.extend([male_index, female_index])
    values.extend([g11_male, g11_female])

    # Add connections for Grade 12
    g12_male = df[f"g12_{strand}_male"].sum()
    g12_female = df[f"g12_{strand}_female"].sum()
    source.extend([grade_12_index, grade_12_index])
    target.extend([male_index, female_index])
    values.extend([g12_male, g12_female])

    # Male & Female to Strands
    source.extend([male_index, female_index])
    target.extend([strand_index, strand_index])
    values.extend([g11_male + g12_male, g11_female + g12_female])

# Create Sankey diagram with **taller layout**
fig = go.Figure(go.Sankey(
    node=dict(
        pad=50,  # Keeps spacing balanced
        thickness=150,  # Adjust thickness of nodes
        line=dict(color="black", width=1.5),
        label=labels,
        color=["lightgray"] * len(labels),
    ),
    link=dict(
        source=source,
        target=target,
        value=values,
        color=["LightSkyBlue" if i % 2 == 0 else "LightPink" for i in range(len(values))],
    )
))

fig.update_layout(
    title_text="Senior High School Student Distribution",
    font_size=18,
    height=700,  # MAKE IT TALLER
    margin=dict(l=150, r=150, t=100, b=100)
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
file_path = "cleaned_data.csv"  # Update this path if needed
df = pd.read_csv(file_path)

# Define strands in the requested order
strands_dict = {
    "acad_-_humss": "HUMSS",
    "acad_-_abm": "ABM",
    "acad_stem": "STEM",
    "acad_gas": "GAS",
    "tvl": "TVL",
    "sports": "SPORTS",
    "arts": "ARTS",
    "acad_pbm": "PBM"
}

# Define nodes
labels = ["Grade 11", "Grade 12", "Male", "Female"] + list(strands_dict.values()) + ["Senior High School"]

# Assign numeric indices
grade_11_index = labels.index("Grade 11")
grade_12_index = labels.index("Grade 12")
male_index = labels.index("Male")
female_index = labels.index("Female")
shs_index = labels.index("Senior High School")  # New SHS Box

# Initialize link data
source = []
target = []
values = []
link_tooltips = []

# Compute total outflow from Grade 11 & 12 to Male & Female
grade_11_male_total = sum(df[f"g11_{strand}_male"].sum() for strand in strands_dict)
grade_11_female_total = sum(df[f"g11_{strand}_female"].sum() for strand in strands_dict)
grade_12_male_total = sum(df[f"g12_{strand}_male"].sum() for strand in strands_dict)
grade_12_female_total = sum(df[f"g12_{strand}_female"].sum() for strand in strands_dict)

# Grade 11 -> Male & Female
source.extend([grade_11_index, grade_11_index])
target.extend([male_index, female_index])
values.extend([grade_11_male_total, grade_11_female_total])

# Tooltip: Separate Grade 11 & 12 Outflows
link_tooltips.append(f"Grade 11 Male count: {grade_11_male_total}")
link_tooltips.append(f"Grade 11 Female count: {grade_11_female_total}")

# Grade 12 -> Male & Female
source.extend([grade_12_index, grade_12_index])
target.extend([male_index, female_index])
values.extend([grade_12_male_total, grade_12_female_total])

# Tooltip for Grade 12
link_tooltips.append(f"Grade 12 Male count: {grade_12_male_total}")
link_tooltips.append(f"Grade 12 Female count: {grade_12_female_total}")

# Male & Female to Strands
strand_tooltips = []
total_shs_students = 0

for strand in strands_dict:
    strand_index = labels.index(strands_dict[strand])

    male_flow = df[f"g11_{strand}_male"].sum() + df[f"g12_{strand}_male"].sum()
    female_flow = df[f"g11_{strand}_female"].sum() + df[f"g12_{strand}_female"].sum()
    total_flow = male_flow + female_flow
    total_shs_students += total_flow

    source.extend([male_index, female_index])
    target.extend([strand_index, strand_index])
    values.extend([male_flow, female_flow])

    # Outflow tooltips for Male & Female -> Strands
    link_tooltips.append(f"Males in {strands_dict[strand]}: {male_flow}")
    link_tooltips.append(f"Females in {strands_dict[strand]}: {female_flow}")

    # Strand box tooltips (Restored Total Counts)
    strand_tooltips.append(f"{strands_dict[strand]}<br>Females: {female_flow}<br>Males: {male_flow}<br>Total: {total_flow}")

# Strands -> Senior High School (Final Box)
for strand in strands_dict:
    strand_index = labels.index(strands_dict[strand])

    strand_total = df[f"g11_{strand}_male"].sum() + df[f"g12_{strand}_male"].sum() + \
                   df[f"g11_{strand}_female"].sum() + df[f"g12_{strand}_female"].sum()

    source.append(strand_index)
    target.append(shs_index)
    values.append(strand_total)

    link_tooltips.append(f"Total students in {strands_dict[strand]}: {strand_total}")

# Node (Box) Tooltips
node_tooltips = [
    f"GRADE 11:<br>Female: {grade_11_female_total}<br>Male: {grade_11_male_total}<br>Total: {grade_11_female_total + grade_11_male_total}",
    f"GRADE 12:<br>Female: {grade_12_female_total}<br>Male: {grade_12_male_total}<br>Total: {grade_12_female_total + grade_12_male_total}",
    f"Male Students in SHS:<br>{grade_11_male_total + grade_12_male_total}",
    f"Female Students in SHS:<br>{grade_11_female_total + grade_12_female_total}",
] + strand_tooltips + [f"SENIOR HIGH SCHOOL:<br>Total Students: {total_shs_students}"]

# Define colors (Male = Blue, Female = Pink, Others = Light Gray, SHS = Dark Gray)
node_colors = [
    "lightgray",  # Grade 11
    "lightgray",  # Grade 12
    "deepskyblue",  # Male
    "lightpink",  # Female
] + ["lightgray"] * len(strands_dict) + ["darkgray"]  # SHS in Dark Gray

fig = go.Figure(go.Sankey(
    node=dict(
        pad=50,
        thickness=100,
        line=dict(color="black", width=1.5),
        label=labels,
        color=node_colors,
        customdata=node_tooltips,
        hovertemplate='%{customdata}<extra></extra>',
    ),
    link=dict(
        source=source,
        target=target,
        value=values,
        color=["LightSkyBlue" if i % 2 == 0 else "LightPink" for i in range(len(values))],
        customdata=link_tooltips,
        hovertemplate='%{customdata}<extra></extra>',
    )
))

fig.update_layout(
    title_text="Senior High School Student Distribution",
    font_size=18,
    height=500,
    margin=dict(l=150, r=150, t=100, b=100)
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load dataset
file_path = "cleaned_data.csv"
df = pd.read_csv(file_path)

# Define categories
levels = {
    "Elementary School": {
        "Kindergarten": ["k_male", "k_female"],
        "Grade 1": ["g1_male", "g1_female"],
        "Grade 2": ["g2_male", "g2_female"],
        "Grade 3": ["g3_male", "g3_female"],
        "Grade 4": ["g4_male", "g4_female"],
        "Grade 5": ["g5_male", "g5_female"],
        "Grade 6": ["g6_male", "g6_female"],
        "Non-graded": ["elem_ng_male", "elem_ng_female"]
    },
    "Junior High School": {
        "Grade 7": ["g7_male", "g7_female"],
        "Grade 8": ["g8_male", "g8_female"],
        "Grade 9": ["g9_male", "g9_female"],
        "Grade 10": ["g10_male", "g10_female"],
        "Non-graded": ["jhs_ng_male", "jhs_ng_female"]
    },
    "Senior High School": {
        "HUMSS": ["g11_acad_-_humss_male", "g11_acad_-_humss_female", "g12_acad_-_humss_male", "g12_acad_-_humss_female"],
        "ABM": ["g11_acad_-_abm_male", "g11_acad_-_abm_female", "g12_acad_-_abm_male", "g12_acad_-_abm_female"],
        "STEM": ["g11_acad_stem_male", "g11_acad_stem_female", "g12_acad_stem_male", "g12_acad_stem_female"],
        "GAS": ["g11_acad_gas_male", "g11_acad_gas_female", "g12_acad_gas_male", "g12_acad_gas_female"],
        "TVL": ["g11_tvl_male", "g11_tvl_female", "g12_tvl_male", "g12_tvl_female"],
        "SPORTS": ["g11_sports_male", "g11_sports_female", "g12_sports_male", "g12_sports_female"],
        "ARTS": ["g11_arts_male", "g11_arts_female", "g12_arts_male", "g12_arts_female"],
        "PBM": ["g11_acad_pbm_male", "g11_acad_pbm_female", "g12_acad_pbm_male", "g12_acad_pbm_female"]
    }
}

fig = go.Figure()
menu_buttons = []

for level, categories in levels.items():
    totals = {}
    for category, columns in categories.items():
        if len(columns) == 2:
            male_total = df[columns[0]].sum()
            female_total = df[columns[1]].sum()
        else:
            male_total = df[columns[0]].sum() + df[columns[2]].sum()
            female_total = df[columns[1]].sum() + df[columns[3]].sum()
        totals[category] = {"Male": male_total, "Female": female_total, "Total": male_total + female_total}

    labels = list(categories.keys()) + ["Male", "Female", level]
    male_index = labels.index("Male")
    female_index = labels.index("Female")
    level_index = labels.index(level)

    source, target, values, link_tooltips, node_tooltips = [], [], [], [], []

    for category, counts in totals.items():
        node_tooltips.append(f"{category}: {counts['Total']}")
    node_tooltips.append(f"Total Male: {sum(totals[cat]['Male'] for cat in totals)}")
    node_tooltips.append(f"Total Female: {sum(totals[cat]['Female'] for cat in totals)}")
    node_tooltips.append(f"Overall Population in {level}: {sum(totals[cat]['Total'] for cat in totals)}")

    for category, counts in totals.items():
        category_index = labels.index(category)
        source.extend([category_index, category_index])
        target.extend([male_index, female_index])
        values.extend([counts["Male"], counts["Female"]])
        link_tooltips.append(f"Males in {category}: {counts['Male']}")
        link_tooltips.append(f"Females in {category}: {counts['Female']}")

    source.extend([male_index, female_index])
    target.extend([level_index, level_index])
    values.extend([
        sum(totals[cat]["Male"] for cat in totals),
        sum(totals[cat]["Female"] for cat in totals),
    ])
    link_tooltips.append(f"Total Count : {values[-2]}")
    link_tooltips.append(f"Total Count : {values[-1]}")

    fig.add_trace(go.Sankey(
        visible=(level == "Elementary School"),
        node=dict(
            pad=50,
            thickness=100,
            line=dict(color="black", width=1.5),
            label=labels,
            color=["lightgray"] * len(categories) + ["deepskyblue", "lightpink", "gray"],
            customdata=node_tooltips,
            hovertemplate='%{customdata}<extra></extra>'
        ),
        link=dict(
            source=source,
            target=target,
            value=values,
            color=["LightSkyBlue" if i % 2 == 0 else "LightPink" for i in range(len(values))],
            customdata=link_tooltips,
            hovertemplate='%{customdata}<extra></extra>'
        )
    ))

    menu_buttons.append(dict(
        label=level,
        method="update",
        args=[{"visible": [level == l for l in levels] + [False]}]
    ))

# Add Overall Distribution
total_male = df.filter(like='_male').sum().sum()
total_female = df.filter(like='_female').sum().sum()
total_students = total_male + total_female

labels_overall = ["Male Students", "Female Students", "Total Students"]
source_overall, target_overall, values_overall = [0, 1], [2, 2], [total_male, total_female]

fig.add_trace(go.Sankey(
    visible=False,
    node=dict(
        pad=50,
        thickness=100,
        line=dict(color="black", width=1.5),
        label=labels_overall,
        color=["deepskyblue", "pink", "gray"]
    ),
    link=dict(
        source=source_overall,
        target=target_overall,
        value=values_overall,
        color=["LightSkyBlue", "pink"]
    )
))

# Add dropdown option for Overall Distribution
menu_buttons.append(dict(
    label="Overall Distribution",
    method="update",
    args=[{"visible": [False] * len(levels) + [True]}]
))

# Update layout styling
fig.update_layout(
    title_text="Gender-Based Student Distribution per Curriculum",
    title_x=0.5,
    font=dict(
        family="Poppins",
        size=14,
        color="#444444"
    ),
    title_font=dict(
        family="Poppins",
        size=14,
        color="#444444",
        weight="bold"
    ),
    margin={"r": 0, "t": 0, "l": 0, "b": 0},
    plot_bgcolor='rgba(0, 0, 0, 0)',
    paper_bgcolor='rgba(0, 0, 0, 0)',
    updatemenus=[dict(
        buttons=menu_buttons,
        direction="down",
        showactive=True,
        x=0.65,
        y=1.1,
        xanchor="center",
        yanchor="top"
    )]
)

fig.show()
```
