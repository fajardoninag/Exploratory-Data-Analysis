```python
# Import necessary libraries
from pyspark import SparkContext, SparkConf

# Create a Spark Context
conf = SparkConf().setAppName("BigDataLab").setMaster("local[*]")
sc = SparkContext.getOrCreate(conf)

# Uploading dataset to the RDD
rdd = sc.textFile("suicide_rates.csv")
row = rdd.map(lambda row_line: row_line.split(","))

# removing the header from the index
header = row.first()
rdd_data = row.filter(lambda line: line != header)


# Define safe_parse_int function
def safe_parse_int(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return 0

# Step 1: Map each record to (year, suicide cases)
year_cases_rdd = rdd_data.map(lambda x: (x[1], safe_parse_int(x[4])))  # x[1] is the year, x[4] is suicide cases

# Step 2: Sum the suicide cases for each year using reduceByKey()
total_cases_per_year = year_cases_rdd.reduceByKey(lambda a, b: a + b)

# Step 3: Sort by year for least cases (ascending)
sorted_by_least_cases = total_cases_per_year.sortBy(lambda x: x[1], ascending=True)

# Step 4: Sort by year for most cases (descending)
sorted_by_most_cases = total_cases_per_year.sortBy(lambda x: x[1], ascending=False)

# collect 12 with the least suicide cases
top_12_least_years = sorted_by_least_cases.take(12)

# collect 12 the most suicide cases
top_12_most_years = sorted_by_most_cases.take(12)

# print 12 with the most suicide cases
print("\nTop 12 years with the most suicide cases:\n")
for i, (year, total_cases) in enumerate(top_12_most_years, 1):
    print(f"{i} {year}\nSuicide cases: {total_cases}")

# print 12 with the least suicide cases
print("\nTop 12 years with the least suicide cases:\n")
for i, (year, total_cases) in enumerate(top_12_least_years, 1):
    print(f"{i} {year}\nSuicide cases: {total_cases}")


```

```python
# Introduction to Big Data: Laboratory 1
# Programmed by: Fajardo, Niña G. (BSCpE 3-3)

# Add the libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum, row_number
from pyspark.sql.window import Window

# Initializing Spark session
spark = SparkSession.builder.appName("SuicideData").getOrCreate()

# Loading CSV file into DataFrame
df_csv = spark.read.csv("suicide_rates.csv", header=True, inferSchema=True)

# Selecting the columns from the CSV file
df_selected = df_csv.select("year", "suicides_no")

# Renaming the columns to make them more descriptive
df_renamed = df_selected.withColumnRenamed("year", "Year") \
                        .withColumnRenamed("suicides_no", "Suicide Cases")

# Group by 'Year' and sum the 'Suicide Cases' for each year
df_grouped_by_year = df_renamed.groupBy("Year").agg(sum("Suicide Cases").alias("Total Suicide Cases"))

# Sorting by descending order to find the years with the most suicide cases
df_most_cases = df_grouped_by_year.orderBy("Total Suicide Cases", ascending=False).limit(12)

# Sorting by ascending order to find the years with the least suicide cases
df_least_cases = df_grouped_by_year.orderBy("Total Suicide Cases", ascending=True).limit(12)

# Define a window specification for row_number
window_spec = Window.orderBy("Total Suicide Cases")

# Adding a row number (ranking) to both DataFrames using row_number
df_most_cases = df_most_cases.withColumn("Number", row_number().over(window_spec))
df_least_cases = df_least_cases.withColumn("Number", row_number().over(window_spec))

# Output
print("Top 12 Years with Most Suicide Cases:")
df_most_cases.select("Number", "Year", "Total Suicide Cases").show(truncate=False)

print("\nTop 12 Years with Least Suicide Cases:")
df_least_cases.select("Number", "Year", "Total Suicide Cases").show(truncate=False)


```

```python
df_csv.createOrReplaceTempView("suicide_rates")

# Find the year with the most suicide deaths
year_most_suicides = spark.sql("""
SELECT year, SUM(suicides_no) AS total_suicides
FROM suicide_rates
GROUP BY year
ORDER BY total_suicides DESC
LIMIT 1
""")
year_most_suicides.show()

# Find the country with the most suicide deaths
country_most_suicides = spark.sql("""
SELECT country, SUM(suicides_no) AS total_suicides
FROM suicide_rates
GROUP BY country
ORDER BY total_suicides DESC
LIMIT 1
""")
country_most_suicides.show()

# Find the age range with the most suicide deaths
age_range_most_suicides = spark.sql("""
SELECT
    CASE
        WHEN age BETWEEN 0 AND 20 THEN '0-20'
        WHEN age BETWEEN 21 AND 30 THEN '21-30'
        WHEN age BETWEEN 31 AND 40 THEN '31-40'
        WHEN age BETWEEN 41 AND 50 THEN '41-50'
        WHEN age BETWEEN 51 AND 60 THEN '51-60'
        WHEN age BETWEEN 61 AND 70 THEN '61-70'
        WHEN age BETWEEN 71 AND 80 THEN '71-80'
        WHEN age BETWEEN 81 AND 90 THEN '81-90'
        WHEN age BETWEEN 91 AND 100 THEN '91-100'
        ELSE '100+'
    END AS age_range,
    SUM(suicides_no) AS total_suicides
FROM suicide_rates
GROUP BY age_range
ORDER BY total_suicides DESC
LIMIT 1
""")
age_range_most_suicides.show()
```
