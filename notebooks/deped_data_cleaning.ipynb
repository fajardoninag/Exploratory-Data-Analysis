```python
import re

# Clean the street address column
def clean_street_address(address):
    if not isinstance(address, str):
        return "Not provided"

    address = address.strip()  # Remove extra spaces

    # Unexpected characters
    address = address.replace('"', '').replace("'", "").strip()

    # Define invalid values to replace with "Not provided"
    missing_values = {"0", "0.0", "not applicable", "Not Applicable",
                      "not available", "Not Available", "-none",
                      "none", "n/a", "N/A", "N / A", "n / a",
                      "_ _ _ _", "______", "-----", "------"}

    if address.lower() in missing_values:
        return "Not provided"

    # If the value is just multiple zeroes, replace it
    if re.fullmatch(r"0+", address):
        return "Not provided"

    # Remove leading "- ", "_", "*", ",", or "." if present
    address = re.sub(r"^[\s,._-]+", "", address)

    # Fix "@nd" → "2nd"
    address = re.sub(r"@nd", "2nd", address, flags=re.IGNORECASE)

    # If less than 5 characters after cleaning, set to "Not provided"
    if len(address) < 5:
        return "Not provided"

    return address

# Apply function to street_address column
df_cleaned['street_address'] = df_cleaned['street_address'].apply(clean_street_address)

# Extra forced replacement for any remaining quoted "0", "N/A", or "None"
df_cleaned['street_address'] = df_cleaned['street_address'].replace(
    to_replace=r'^\s*["\']?(0+|N/A|None|none|n/a|not available|not applicable)["\']?\s*$',
    value="Not provided",
    regex=True
)
```

```python
# Count 0s
zero_values_count = (df_cleaned['street_address'] == "0").sum()

print(f'Number of "0" values in street_address: {zero_values_count}')
```

```python
# Trim spaces that are double, trailing, and leading
def trim_whitespace(value):
    if isinstance(value, str):
        return " ".join(value.split())
    return value

# Columns 1 to 14
df_cleaned.iloc[:, 0:14] = df_cleaned.iloc[:, 0:14].apply(lambda col: col.apply(trim_whitespace) if col.dtype == "object" else col)
```

```python
# Consistent format in division column
def clean_division(value):
    if isinstance(value, str):
        if value.lower().startswith("city of "):
            return value[8:] + " City"

        # Fix Lanao del Sur - I, inconsistent formart
        if " - " in value:
            value = value.replace(" - ", " ")

        # To match format with other Samar regions
        if value.lower() == "samar (western samar)":
            return "Western Samar"

    return value
```

```python
df_cleaned['division'] = df_cleaned['division'].apply(clean_division)
```

```python
# Remove spaces around hyphens in district names
def clean_district(value):
    if isinstance(value, str):
        value = value.strip()
        value = value.replace(" - ", "-")
    return value

```

```python
df_cleaned['district'] = df_cleaned['district'].apply(clean_district)
```

```python
import re

# Function to properly capitalize each word, including after hyphens and parentheses
def capitalize_properly(value):
    if isinstance(value, str):

        return re.sub(r"(\b[a-z])|(?<=\(|-)[a-z]", lambda x: x.group().upper(), value.lower())
    return value
```

```python
# Apply function to province, municipality, and barangay columns
df_cleaned['province'] = df_cleaned['province'].apply(capitalize_properly)
df_cleaned['municipality'] = df_cleaned['municipality'].apply(capitalize_properly)
df_cleaned['barangay'] = df_cleaned['barangay'].apply(capitalize_properly)
```
